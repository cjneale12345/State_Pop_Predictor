```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from scipy import stats
```

```python
colnames = ['state', 'ages', 'year', 'population']
```

```python
df = pd.read_csv('state-population.csv', names=colnames, header=0)
```

#State Selection

<i>Below type in the two letter code for the state that you would like to model</i>

```python
df2 = df.loc[(df['state'] == 'KY') & (df['ages'] == 'total')]
```

--End State Selection--

```python
x = np.array(df2['year'])
```

```python
y = np.array(df2['population'])
```

```python
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
def myfunc(x):
  return slope * x + intercept
```

```python
mymodel = list(map(myfunc, x))
```

```python
plt.plot(x, mymodel)
```

```python
plt.scatter(x, y)
```

```python
plt.show()
```

#Prediction Function
<i>Below input the year that you would like to predict for</i>

```python
year = myfunc(2030)
```
